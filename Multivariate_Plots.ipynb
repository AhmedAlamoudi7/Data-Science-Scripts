```python
from google.colab import drive
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

drive.mount('/content/drive')

# Update with the path that matches your file's location inside Drive
file_path = '/content/drive/My Drive/week 2/data-cleaning-core.csv'
df = pd.read_csv(file_path)
df.head()
```

1️⃣ Plot — Number of Passengers by Sex, Colored by Survival

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Sex", hue="Survived", palette="coolwarm")
plt.title("Passenger Count by Sex and Survival Status")
plt.xlabel("Sex")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No (0)", "Yes (1)"])
plt.show()
```

2️⃣ Plot — Average Number of “Parents or Children” by Survival

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Survived", y="Parents or Children", palette="Set2")
plt.title("Average Number of Parents/Children by Survival")
plt.xlabel("Survived (0 = Died, 1 = Survived)")
plt.ylabel("Average Parents or Children")
plt.show()
```

3️⃣ Plot — Correlation Heatmap

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Columns")
plt.show()
```
