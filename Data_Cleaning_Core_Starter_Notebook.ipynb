## G) Missing Values

````python
# Get the number of null values for all columns
null_counts = df_titanic.isnull().sum()

# Display the result
null_counts
````

````python
# Calculate percentage of null values for each column
null_percentage = df_titanic.isnull().mean() * 100

# Display the result
null_percentage
````

````python
# Import the package
import missingno as msno

# Visualize missing values with a matrix plot
msno.matrix(df_titanic)
````

#### Address the null values:

- Fill the null values in categorical columns with a placeholder category called MISSING.


````python
# Get a list of categorical columns
categorical_columns = df_titanic.select_dtypes(include='object').columns.tolist()

# Display the list
categorical_columns
````

````python
# Fill null values in categorical columns with 'MISSING'
df_titanic[categorical_columns] = df_titanic[categorical_columns].fillna('MISSING')

# Check to confirm
df_titanic[categorical_columns].isnull().sum()
````


- Fill the null values in numeric columns with a placeholder value of -1



````python
# Get a list of numeric columns
numeric_columns = df_titanic.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Display the list
numeric_columns
````

````python
# Fill null values in numeric columns with -1
df_titanic[numeric_columns] = df_titanic[numeric_columns].fillna(-1)

# Check to confirm
df_titanic[numeric_columns].isnull().sum()
````

- Use either missingno's matrix plot or Pandas to confirm that there are 0 null values remaining.
